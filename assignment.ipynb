```python
import numpy as np

np.random.seed(0)  
months = np.arange(1, 13) 
sales = 3000 + 200 * months + np.random.normal(0, 300, 12)  
advertising_costs = 500 + 50 * months + np.random.normal(0, 100, 12) 

print("Month\tSales\tAdvertising Cost")
for month, sale, cost in zip(months, sales, advertising_costs):
    print(f"{month}\t{sale:.2f}\t{cost:.2f}")
    import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

months = months.reshape(-1, 1)
sales = sales.reshape(-1, 1)
advertising_costs = advertising_costs.reshape(-1, 1)

model = LinearRegression()
model.fit(advertising_costs, sales)

sales_pred = model.predict(advertising_costs)

plt.scatter(advertising_costs, sales, label='Actual Data')
plt.plot(advertising_costs, sales_pred, color='red', label='Linear Regression')
plt.xlabel('Advertising Cost')
plt.ylabel('Sales')
plt.title('Relationship between Sales and Advertising Cost')
plt.legend()
plt.show()

print('Slope (Coefficient):', model.coef_[0][0])
print('Intercept:', model.intercept_[0])
```
